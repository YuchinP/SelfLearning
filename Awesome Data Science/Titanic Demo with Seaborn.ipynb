```python
#import libraries
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib as plt
import seaborn as sns
```

```python
#enable inline plots
%matplotlib inline
```

```python
#pandas reads .csv of passengers on titanic and store in a dataframe
titanic = pd.read_csv('train.csv')
```

```python
#first few lines
titanic.head()
```

```python
#basic information about the dataframe
titanic.info()
```

```python
#get counts by sex or gender
titanic['Sex'].value_counts()
```

```python
# get counts by survival
titanic['Survived'].value_counts()
```

```python
sns.factorplot('Survived', data=titanic,kind='count')
```

```python
#survivability by cabin class
sns.factorplot('Pclass','Survived', data = titanic)
```

```python
# survivability by sex
sns.factorplot('Sex', 'Survived', data = titanic)
```

```python
#plot survivability by sex and broken down by cabin class
sns.factorplot('Sex', 'Survived', hue='Pclass', data=titanic)
```

```python
#plot histogram of age
titanic['Age'].hist(bins=50)
```

```python
sns.distplot(titanic['Age'].dropna())
```

```python
#use facet grid to plot multiple kde plots of age and sex on one plot
fig = sns.FacetGrid(titanic, hue='Sex', aspect = 4)
fig.map(sns.kdeplot,'Age',shade=True)
oldest = titanic['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
#use facet grid to plot multiple kdeplots of age and pclass on one plot
fig = sns.FacetGrid(titanic,hue='Pclass',aspect=4)
fig.map(sns.kdeplot,'Age', shade='True')
oldest= titanic['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```
